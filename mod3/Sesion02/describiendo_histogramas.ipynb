```python
import pandas as pd
import seaborn as sns
from scipy.stats import kurtosis, skew
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/beduExpert/B2-Analisis-de-Datos-con-Python-2020-Santander/main/Datasets/diabetes-clean.csv", index_col=0)
indeps = df.drop("outcome", axis=1)
nonZero = ["glucose", "blood_pressure", "skin_thickness", "insulin", "bmi"]
```

Before cleaning no-sense zeros

```python
for col in indeps.columns:
    print(f"\nAnalising column {col}...")
    data = indeps[col]
    sns.distplot(data, kde=False, norm_hist=False, bins=50)
    plt.show()
    print(f"Kurtosis: {kurtosis(data)}\tSkew: {skew(data)}")
```

After cleaning non-sense zeros

```python
dt = indeps["insulin"]
print(dt.count())
dt[dt == 0].count()
```

```python
for col in indeps.columns:
    print(f"\nAnalising column {col}...")
    data = indeps[col]
    if col in nonZero:
        data = data[data != 0]
    sns.distplot(data, kde=False, norm_hist=False, bins=50)
    plt.show()
    print(f"Kurtosis: {kurtosis(data)}\tSkew: {skew(data)}")
```
