```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/beduExpert/B2-Analisis-de-Datos-con-Python-2020-Santander/main/Datasets/diabetes-clean.csv', index_col=0)

df.head()
```

### a)

```python
sns.distplot(df['glucose'], kde=False, norm_hist=False)
plt.title("Distribución niveles de glucosa en sangre")
plt.xlabel("Glucosa")
plt.ylabel("Recuento de personas");
```

### b)

```python
pregnancy_counts = df['pregnancies'].value_counts()
```

```python
sns.barplot(pregnancy_counts.index, pregnancy_counts)
plt.title("Recuento número de embarazos por mujer")
plt.xlabel("Número de embarazos")
plt.ylabel("Recuento");
```

### c)

```python
healthy = df[df['outcome'] == 0]
sick = df[df['outcome'] == 1]
```

```python
sns.distplot(healthy['glucose'], hist=False, kde_kws={'shade': True})
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True})
plt.title("Comparativo nivel de glucosa entre mujeres con y sin diabetes")
plt.xlabel("Glucosa")
plt.ylabel("Fracción de aparición")
plt.legend(["Sanas", "Enfermas"]);
```

### d)

```python
sns.distplot(sick['blood_pressure'], hist=False, kde_kws={'shade': True})
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True})
sns.distplot(sick['insulin'], hist=False, kde_kws={'shade': True})
plt.title("Distribución de niveles relevantes en enfermas de diabetes")
plt.xlabel("Nivel")
plt.ylabel("Fracción de incidencia");
plt.legend(["Presión sanguínea", "Glucosa", "Insulina"])
```
